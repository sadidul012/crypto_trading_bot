```python
import pandas as pd

from process.data import load_data
```

```python
df["date"] = df.d.apply(lambda x: pd.to_datetime(x).strftime('%Y/%m/%d'))
df
```

```python
df.groupby("date").agg({
    "o": "first",
    "h": "max",
    "l": "min",
    "c": "last",
    "v": "sum",
})
```
